Relation between the **Price and Minimum Nights** .
In this part we are going to plot a boxplot to see how the price effects the duration of a person's stay at the AirBnB.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
from Scripts import project_functions
df=project_functions.load_and_process('../../Data/raw/listings.csv')
df
sns.boxplot(x = 'minimum_nights', y = 'price', data = df) 
sns.set(rc={'figure.figsize':(15.7,12.27)})
```

According to the trend , With increase in the stay of duration the number of people willing to pay greater price decreases. Most of the people who stay for a short period of time often pay more money as compared to people who prefer to stay for longer periods. The price range of 40000 and above could only be seen in the cases where people are spending less nights at the BnB with a few exceptions.

Relation between **Price** and **The type of Room** .


```python
df.boxplot(column='price', by='room_type')
```

We plot a boxplot to observe what range of price are people willing to pay for different types of room. We have four types of accomadations available for the customers. In addition to that people are willing to pay a greater price for the entire place as compared to other types of accomadations available.
